```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Path to dataset
data_path = '../data/processed/dataset'

# Get class names
classes = os.listdir(data_path)
classes = [c for c in classes if os.path.isdir(os.path.join(data_path, c))]

print(f"Found {len(classes)} classes:")
for cls in classes:
    num_images = len(os.listdir(os.path.join(data_path, cls)))
    print(f"  - {cls}: {num_images} images")
```

```python
import random

# Visualize sample images from each class
fig, axes = plt.subplots(4, 4, figsize=(12, 12))
fig.suptitle('Sample Images from Each Class', fontsize=16)

for i, cls in enumerate(classes):
    cls_path = os.path.join(data_path, cls)
    images = os.listdir(cls_path)
    
    # Select 4 random images
    sample_images = random.sample(images, 4)
    
    for j, img_name in enumerate(sample_images):
        img_path = os.path.join(cls_path, img_name)
        img = Image.open(img_path)
        
        axes[i, j].imshow(img)
        axes[i, j].axis('off')
        if j == 0:
            axes[i, j].set_title(cls, fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```
